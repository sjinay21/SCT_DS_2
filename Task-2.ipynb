### Task 2:Data Cleaning & Exploratory Data Analysis (EDA) on Titanic Dataset

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set visualization style
sns.set_style("darkgrid")
```

```python
# Load train dataset
train_path = "train.csv"
train_df = pd.read_csv(train_path)

# Display the first few rows
print("Train Dataset Preview:")
display(train_df.head())

# Basic information about the train dataset
print("\nTrain Dataset Info:")
print(train_df.info())

# Checking missing values in train dataset
print("\nMissing Values in Train Dataset:")
print(train_df.isnull().sum())
```

```python
# Load test dataset
test_path = "test.csv"
test_df = pd.read_csv(test_path)

# Display the first few rows
print("Test Dataset Preview:")
display(test_df.head())

# Basic information about the test dataset
print("\nTest Dataset Info:")
print(test_df.info())

# Checking missing values in test dataset
print("\nMissing Values in Test Dataset:")
print(test_df.isnull().sum())
```

```python
# Check missing values in Train data
print("\nMissing values in Train Data:")
print(train_df.isnull().sum())

# Check missing values in Test data
print("\nMissing values in Test Data:")
print(test_df.isnull().sum())
```

```python
sns.countplot(data=train_df, x="Survived", hue="Survived", palette="coolwarm", legend=False)
plt.title("Survival Distribution")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Sex", data=train_df, palette="coolwarm")
plt.title("Survival Count by Gender")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Pclass", data=train_df, palette="coolwarm")
plt.title("Survival Count by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(train_df["Age"], bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Rename 'Sex' to 'Gender' for clarity
train_df.rename(columns={'Sex': 'Gender'}, inplace=True)
test_df.rename(columns={'Sex': 'Gender'}, inplace=True)

# Initialize encoders
gender_encoder = LabelEncoder()
embarked_encoder = LabelEncoder()

# Encode Gender (safe)
train_df['Gender'] = gender_encoder.fit_transform(train_df['Gender'].astype(str))
test_df['Gender'] = gender_encoder.transform(test_df['Gender'].astype(str))

# Ensure Embarked is string and fill missing values (avoid inplace warning)
train_df['Embarked'] = train_df['Embarked'].astype(str).fillna('S')
test_df['Embarked'] = test_df['Embarked'].astype(str).fillna('S')

# Encode Embarked (safe)
train_df['Embarked'] = embarked_encoder.fit_transform(train_df['Embarked'])
test_df['Embarked'] = embarked_encoder.transform(test_df['Embarked'])

# Preview the datasets
print("\nPreview of Train Dataset after Encoding:")
display(train_df.head())

print("\nPreview of Test Dataset after Encoding:")
display(test_df.head())
```

```python
train_df.to_csv("cleaned_train.csv", index=False)
test_df.to_csv("cleaned_test.csv", index=False)
print(" Cleaned datasets saved successfully!")
```
